```python
#barchart
import matplotlib.pyplot as plt
lang=['Java','C','PHP','Python']
pop=[22.6,17.6,8.8,7.7]
plt.bar(lang,pop)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("popularity of languages")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('data.csv')
df.plot.bar()
plt.figure(figsize=(10,7))
plt.show()
```

```python
df=pd.read_csv('iris.csv')
import matplotlib.pyplot as plt
plt.scatter('petal_length','petal_width',
           s='sepal_length',alpha=0.5,
            data=df)
plt.xlabel('petal_length',size=16)
plt.ylabel('petal_length',size=16)
plt.title("bubble plot",size=18)
```

```python
import seaborn as sns
iris=sns.load_dataset('iris')
```

```python
#join plot
sns.jointplot(x='sepal_length',y="sepal_width",data=iris,height=4)
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
```

```python
sns.boxplot(data=iris)
```

```python
#pairplot
sns.pairplot(data=iris,kind='scatter')
```

```python
sns.heatmap(iris.corr(),annot=True)
```
